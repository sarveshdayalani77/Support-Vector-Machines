```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.svm import SVC
from sklearn.model_selection import train_test_split,cross_val_score

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
forest = pd.read_csv('/content/forestfires (2).csv')
forest
```

```python
data = forest.describe
```

```python
forest.drop(["month","day"],axis=1,inplace =True)
```

```python
pred = forest.iloc[:,0:28]
target = forest.iloc[:,28]
```

```python
def norm_func(i):
  x = (i-i.min()) / (i.max()-i.min())
  return (x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(pred,target,test_size = 0.25, stratify = target)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

**Linear Kernel**

```python
model = SVC(kernel = 'linear')
model.fit(x_train , y_train)
```

```python
pred_test = model.predict(x_test)
pred_test
```

```python
np.mean(pred_test==y_test)
```

```python
acc = accuracy_score(y_test, pred_test) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, pred_test)
```

**Poly Kernel**

```python
model_poly = SVC(kernel = "poly")
model_poly.fit(x_train,y_train)
```

```python
pred_test_poly = model_poly.predict(x_test)
np.mean(pred_test_poly==y_test)
```

**RBF Kernel**

```python
model_rbf = SVC(kernel = "rbf")
model_rbf.fit(x_train,y_train)
```

```python
pred_test_rbf = model_rbf.predict(x_test)
np.mean(pred_test_rbf==y_test)
```

**Sigmoid Kernel**

```python
model_sig = SVC(kernel = "sigmoid")
model_sig.fit(x_train,y_train)
pred_test_sig = model_rbf.predict(x_test)
np.mean(pred_test_sig==y_test)
```
